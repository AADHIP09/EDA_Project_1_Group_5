```python
import pandas as pd
from pathlib import Path
```

```python
car_sales1 = Path('../csv_data/car_sales_2021_df.csv')
car_sales2 = Path('../csv_data/Imputed_data.csv')
```

```python
car_df1 = pd.read_excel(car_sales1)
car_df2 = pd.read_csv(car_sales2)
```

```python
# Combine the data into a single dataset.  
merged_df = pd.merge(car_df1, car_df2, how="left", on=["Model", "Model"])
merged_df
```

```python
merged_df.to_csv("../csv_data/editable_data.csv")
```

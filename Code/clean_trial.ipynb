```python
import pandas as pd
from pathlib import Path
import numpy as np
from sklearn.impute import SimpleImputer 
```

```python
data = Path("../csv_data/new_final_data.csv")

new_df = pd.read_csv(data)
new_df
```

```python
value_df = new_df.fillna(-1)
value_df
```

```python
imputer = SimpleImputer(missing_values=-1.0, strategy='mean')
imputer = imputer.fit(value_df[['MSRP','Invoice','Sum']])

imputed_df = value_df.copy()
imputed_df[['MSRP','Invoice','Sum']] = imputer.transform(value_df[['MSRP','Invoice','Sum']])
imputed_df
```

```python
new_imp_df = imputed_df.replace(to_replace=-1, value=np.NaN)
new_imp_df
```

```python
new_imp_df.to_csv("../csv_data/Imputed_data.csv")
```

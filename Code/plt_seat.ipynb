```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import numpy as np
from scipy import stats
```

```python
plt_seat_df = Path("../Output/seat_data.csv")
df = pd.read_csv(plt_seat_df)
df.head()
```

```python
a, b = np.polyfit(df['power_ps'], df['price_in_euro'], 1)
```

```python
plt.scatter(df['power_ps'], df['price_in_euro'], marker="o", facecolors="red", edgecolors="black",
            alpha=0.5)
plt.xlabel("power(ps)")
plt.ylabel("price(euro)")
plt.title("Seat Power and Price Relations")
plt.plot(df['power_ps'], a*df['power_ps']+b)
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('base_clean/base_data.csv')
df.head()
```

```python
df.info()
```

```python
df.loc[:, 'year'] = df['year'].astype(str)
```

```python
plt.scatter(df['year'], df['price_in_euro'],marker="o")
plt.title('Car Prices Over the Years')
plt.xlabel('Year')
plt.xticks(rotation="vertical")
plt.ylabel('Car Price (Euro)')
plt.show()
```

```python
yearly_avg_prices = df.groupby('year')['price_in_euro'].mean()
yearly_avg_prices
```

```python
plt.figure(figsize=(12, 6))
yearly_avg_prices.plot(marker='o')
plt.title('Average Car Prices Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price (Euro)')
plt.show()
```

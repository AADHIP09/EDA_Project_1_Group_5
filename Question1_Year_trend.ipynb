```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_all = pd.read_csv('base_clean/base_data.csv')
df_Five = pd.read_csv('base_clean/base_5_data.csv')
df_all.head()
```

```python
df_all.info()
```

```python
df_all.loc[:, 'year'] = df_all['year'].astype(str)
df_Five.loc[:, 'year'] = df_Five['year'].astype(str)
```

```python
plt.scatter(df_all['year'], df_all['price_in_euro'],marker="o")
plt.title('Car Prices Over the Years')
plt.xlabel('Year')
plt.xticks(rotation="vertical")
plt.ylabel('Car Price (Euro)')
plt.show()
```

```python
yearly_avg_prices = df_all.groupby('year')['price_in_euro'].mean()
yearly_avg_prices
```

```python
plt.figure(figsize=(12, 6))
yearly_avg_prices.plot(marker='o')
plt.title('Average Car Prices Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price (Euro)')
plt.show()
```

```python
yearly_avg_prices_5 = df_Five.groupby('year')['price_in_euro'].mean()
plt.figure(figsize=(12, 6))
yearly_avg_prices_5.plot(marker='o')
plt.title('Average Car Prices Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price (Euro)')
plt.show()
```

```python
yearly_sales = df_all['year'].value_counts()
yearly_sales = yearly_sales.sort_index(ascending=True)
```

```python
plt.figure(figsize=(12, 6))
yearly_sales.plot(marker='o')
plt.title('Car Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Car Sales')
plt.show()
```
